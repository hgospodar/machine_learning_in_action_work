```python
# 2.2.2使用Matplotlib 创建散点图
import matplotlib.pyplot as plt
import numpy as np
import kNN
import importlib
%matplotlib inline
datingDataMat, datingLbels = kNN.file2matrix('datingTestSet.txt')
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(datingDataMat[:,1], datingDataMat[:,2])
plt.show()
```

```python
# 2.2.2区分数据点属性类别
from numpy import *
fig, ax = plt.subplots()
ax.scatter(datingDataMat[:,1], datingDataMat[:,2],
           15.0*array(datingLbels), 15.0*array(datingLbels))
plt.show()
```

```python
# 2.2.2 切换不同的特征显示，加入图例legend
fig, ax = plt.subplots()
love_dictionary = {'largeDoese':3, 'smallDoeses':2, 'didntLike':1}
color_dictionary = {'largeDoese':'red', 'smallDoeses':'blue', 'didntLike':'green'}
for k in love_dictionary:
    x1 = datingDataMat[array(datingLbels)==love_dictionary[k],0]
    x2 = datingDataMat[array(datingLbels)==love_dictionary[k],1]
    length = shape(x1)[0]
    size = full((length,1), 15.0*love_dictionary[k])
    color = full((length, 1), 15.0*love_dictionary[k])
    ax.scatter(x1, x2, size, color=color_dictionary[k], label = k)
ax.legend()
plt.show()
```
